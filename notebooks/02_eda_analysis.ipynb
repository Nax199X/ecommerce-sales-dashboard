```python
import pandas as pd

df = pd.read_csv("../output/cleaned_data.csv", parse_dates=["invoice_date"])
df.head()
```

```python
df["description"] = df["description"].str.title()
df["description"].head()
```

```python
df.head()
```

```python
total_revenue = df["sales"].sum()
print(total_revenue)
```

```python
total_orders = df["invoice_no"].nunique()
print(total_orders)
```

```python
total_customers = df["customer_id"].nunique()
print(total_customers)
```

```python
df["year"] = df["invoice_date"].dt.year
df["month"] = df["invoice_date"].dt.month
df["month_name"] = df["invoice_date"].dt.month_name()
df["day_of_week"] = df["invoice_date"].dt.day_name()
```

```python
monthly_sales = df.groupby("month_name")["sales"].sum().sort_values()
monthly_sales
```

```python
country_sales = df.groupby("country")["sales"].sum().sort_values(ascending=False)
country_sales.head(10)
```

```python
top_products = df.groupby("description")["sales"].sum().sort_values(ascending=False).head(10)
top_products
```

```python
customer_revenue = df.groupby("customer_id")["sales"].sum().sort_values(ascending=False)
customer_revenue.head()
```

```python
df.to_csv("../output/cleaned_data.csv", index=False)
```
