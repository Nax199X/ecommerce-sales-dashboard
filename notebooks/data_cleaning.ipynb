```python
# Check shape before
print("Before cleaning:", df.shape)

# Remove rows with missing CustomerID
df = df[df["CustomerID"].notna()]

# Remove rows with missing Description
df = df[df["Description"].notna()]

print("After cleaning:", df.shape)
```

```python
df.shape
df.isnull().sum()
df.head()
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df = df.rename(columns={
    "invoiceno": "invoice_no",
    "stockcode": "stock_code",
    "invoicedate": "invoice_date",
    "unitprice": "unit_price",
    "customerid": "customer_id"
})
df.columns
```

```python
df.dtypes
```

```python
df["invoice_date"] = pd.to_datetime(df["invoice_date"])
df["customer_id"] = df["customer_id"].astype(int)
df.dtypes
```

```python
df[df["quantity"] <= 0].head()
```

```python
df[df["invoice_no"].str.startswith("C", na=False)].head()
```

df["sales"] = df["quantity"] * df["unit_price"]
we have too many negative value in quanity column and that will be return . so we will calculate positive value for sale column that for calculating revenue, we will make copy revenue for negative quanity items for later analysis. 


```python
df_sales = df[(df["quantity"] > 0) & (df["unit_price"] > 0)].copy()
df_sales["sales"] = df_sales["quantity"] * df_sales["unit_price"]
df_sales.head()

```

```python
df_sales.to_csv("../output/cleaned_data.csv", index=False)
print("Cleaned data saved to '../output/cleaned_data.csv'")
```

```python
df_returns = df[df["quantity"] < 0].copy()
df_returns["sales"] = df_returns["quantity"] * df_returns["unit_price"]
df_returns.head()
```

```python
df_clean = df[(df["quantity"] > 0) & (df["unit_price"] > 0)].copy()
df_clean["sales"] = df_clean["quantity"] * df_clean["unit_price"]
df_clean.head()
```

```python
df["year"] = df["invoice_date"].dt.year
df["month"] = df["invoice_date"].dt.month
```
